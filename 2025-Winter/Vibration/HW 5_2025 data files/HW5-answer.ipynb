```python
from numpy import (array, pi, sin, cos, linspace, where, array, exp, sqrt, vstack,
diag, sort, dot)
from numpy.linalg import eig, inv, det, norm
from pandas import read_excel
from printarray import printarray
```

```python
KDF = read_excel('stiffness_matrix.xlsx')
MDF = read_excel('mass_matrix.xlsx')
```

```python
K = KDF.values
M = MDF.values
```

```python
A = inv(M)@K
lambda_A,U = eig(A)
omega = sqrt(lambda_A)
msiz = len(lambda_A)

printarray(lambda_A,'.4f',name = 'Eigenvalues lambda_A')
printarray(omega,'.4f',name = 'Natural circular frequencies',eu = 'rad/s')

for ii in range(msiz):
    print('Eigenvector {}'.format(ii+1))
    printarray(U[:,ii],'.4f')
```

```python
omega_sorted = omega.copy()
omega_sorted.sort()
print('Natural circular frequencies sorted in increasing value:')
printarray(omega_sorted,'.4f',eu = 'rad/s')
```

0.0.1 Lowest natural circular frequency

```python
indx = where(omega==omega_sorted[0]) 
# index of the lowest natural circular
# frequency in eigenvector array
indx
```

```python
indxl = indx[0][0]
indxl

print('lowest natural circular frequency = {:.4f} rad/s'
.format(omega[indxl]))
```

Ratio of the first displacement to the last displacement of the mode shape
associated with the lowest natural frequency

```python
U[0,indxl]/U[-1,indxl]
U[0,indxl]/U[7,indxl]
```

Highest natural circular frequency

```python
indx = where(omega==omega_sorted[-1]) 
# index of the highest natural circular
# frequency in eigenvector array
indx
```

```python
indxh = indx[0][0]
indxh
```

```python
print('highest natural circular frequency = {:.2f} rad/s'.format(omega[indxh]))
```

Dot product of the lowest and highest egenvectors

```python
dot(U[:,indxl],U[:,indxh])
```
