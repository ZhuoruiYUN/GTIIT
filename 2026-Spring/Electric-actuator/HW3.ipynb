```python
from numpy import exp
import matplotlib.pyplot as plt
#rotor modeling
#plot the system transient response without load,with 0.1 Nm load and with 0.2 Nm load
#omega = [(KV-T*R_A)/K ^2  - [(KV-T*R_A)]/K^2 * exp(-t/tau)
V = 24
K = 0.0256
J = 14.9e-7
R_A = 2.14
tau = J * R_A / K**2
T = [0, 0.1, 0.2]
t = [i * 0.00003 for i in range(1000)]
omega = []
for i in T:
    omega.append([(K*V - i * R_A) / K**2 - ((K*V - i * R_A) / K**2) * exp(-t[j] / tau) for j in range(len(t))])
    
t_ms = [time * 1000 for time in t]
figuresize = (10, 4)
plt.figure(figsize=figuresize)
plt.plot(t_ms, omega[0], label='No Load')
plt.plot(t_ms, omega[1], label='0.1 Nm Load')
plt.plot(t_ms, omega[2], label='0.2 Nm Load')
plt.xlabel('Time (ms)')
plt.ylabel('Angular Velocity (rad/s)')
plt.title('Transient Response of Rotor')
plt.legend()
plt.grid()
plt.show()

```

```python
# plot the steady state rotation speed ⍵ (in rad/s) vs. torque load (in Nm).
#omega = V/K - T_load*R_A/K^2
torque_load = [0, 0.1, 0.2]
steady_state_omega = [V / K - T * R_A / K**2 for T in torque_load]
torque_load_mNm = [T * 1000 for T in torque_load]
plt.figure(figsize=figuresize)
plt.plot(torque_load_mNm, steady_state_omega, color='b')
plt.scatter(torque_load_mNm, steady_state_omega, color='r')
plt.xlabel('Torque Load (mNm)')
plt.ylabel('Steady State Angular Velocity (rad/s)')
plt.title('Steady State Angular Velocity vs. Torque Load')
plt.grid()
plt.show()
```
